```python
import io
import json
import numpy as np
import tensorflow as tf
```

```python
with open('/kaggle/input/tokenizers/en_output_tokenizer.json') as f:
    data = json.load(f)
    en_tokenizer = tf.keras.preprocessing.text.tokenizer_from_json(data)

with open('/kaggle/input/tokenizers/fr_input_tokenizer.json') as f:
    data = json.load(f)
    fr_tokenizer = tf.keras.preprocessing.text.tokenizer_from_json(data)
```

```python
model = tf.keras.models.load_model('/kaggle/input/translation-models/french-english-translation-model/french-english-translation-model')
```

```python
def translate(input_sentence, input_tokenizer, output_tokenizer, model):
    input_sequence = np.array(input_tokenizer.texts_to_sequences([input_sentence]))
    decoder_inputs = np.array([[output_tokenizer.word_index['<sos>']]])
    while decoder_inputs[0][-1] != output_tokenizer.word_index['<eos>']:
        y_proba = model.predict([input_sequence, decoder_inputs], verbose=0)
        pred = np.argmax(np.squeeze(y_proba, axis=0), 1)
        decoder_inputs = np.append(np.array([output_tokenizer.word_index['<sos>']]), pred)
        decoder_inputs = decoder_inputs.reshape(1, decoder_inputs.shape[0])
    decoder_output = decoder_inputs[0][1:-1]
    output_sentence = ''
    for i in decoder_output:
        output_sentence += output_tokenizer.index_word[i] + ' '
    return output_sentence
```

```python
input_sentence = input("Input French Text: ")
print("Output English Text:", translate(input_sentence, fr_tokenizer, en_tokenizer, model)
```
